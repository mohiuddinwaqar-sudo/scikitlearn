```python
import sklearn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
X, y = iris.data, iris.target

```

```python
sns.set(style="whitegrid")
```

```python
iris["target"]
```

```python
## What problem are we trying to solves? 

#we are trying to classify iris flowers into three species based on four features: sepal length, sepal width, petal length, and petal width.
```

```python
iris_2 = pd.DataFrame(data= iris['data'],columns= iris['feature_names'] )

iris_2['target'] = iris['target']
iris_2.head(10)
```

```python
iris_2.describe()
```

```python
iris_2["sepal length (cm)"].hist(bins=20)
#distribution line

plt.title("Sepal Length Distribution")
```

```python
iris_2["sepal width (cm)"].hist(bins=20)
```

```python
iris_2["petal width (cm)"].hist(bins=20)
```

```python
iris_2["petal length (cm)"].hist(bins=20)
```

```python
iris_2["target_name"]= iris_2["target"].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
```

```python
iris_2.head(10)
```

```python
col = 'sepal length (cm)'
sns.relplot(x=col, y='target', data=iris_2, palette='Set1')
plt.title(f'{col} vs Target Species')
plt.xlabel(col)
plt.ylabel('Target Species')
plt.legend(title='Species', loc='upper right', bbox_to_anchor=(1.5, 1))
plt.show()  
```

```python
sns.pairplot(iris_2, hue = "target_name", palette='Set2')
plt.suptitle('Pairplot of Iris Features Colored by Species', y=1.02)
plt.show()
```

```python
#train test split 
from sklearn.model_selection import train_test_split

iris_train, iris_test = train_test_split(iris_2, test_size = 0.25)

iris_train.shape
iris_test.shape
```

```python
# Prepare our data for modelling 

X_train = iris_train.drop(columns=["target", "target_name"]).values
y_train = iris_train["target"].values
```

```python
y_train
```

### Using a manual modelling approach 

```python
def single_feature_function(petal_length):
    if petal_length < 2.5:
        return 0  # Setosa
    elif petal_length < 4.8:
        return 1  # Versicolor
    else:
        return 2  # Virginica                                        

manual_y_prediction = np.array([single_feature_function(val) for val in X_train[:, 2]])
```

```python
np.mean(manual_y_prediction == y_train)
```

### Using Logistic regression 


```python
#Modelling logistic regression 
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

```

### Using a Validation Set 

```python
Xt, Xv, yt, yv = train_test_split(X_train, y_train, test_size=0.25)
```

```python
model.fit(Xt, yt)
```

```python
y_pred = model.predict(Xv)

np.mean(y_pred == yv)
```

```python
# Testing cross validation
from sklearn.model_selection import cross_val_score, cross_val_predict

model2 = LogisticRegression(max_iter=200)
```

```python
accuracies = cross_val_score(model2, X_train, y_train, cv=5, scoring='accuracy')
np.mean(accuracies)
```

## Where are we misclassifying?

```python
y_pred = cross_val_predict(model2, X_train, y_train, cv=5)

predicted_correctly_mask = y_pred == y_train

predicted_correctly_mask
```

```python
df_predictions = iris_train.copy()
df_predictions["correctly_predicted"] = predicted_correctly_mask
df_predictions.head(10)
```

```python
df_predictions["predictions"] = y_pred
df_predictions["prediction_label"] = df_predictions["predictions"].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
df_predictions.head()
```

```python
sns.scatterplot(data=df_predictions, x="petal length (cm)", y="petal width (cm)", hue="prediction_label", style="correctly_predicted", palette='Set1')
plt.title("Logistic Regression Predictions with Cross-Validation")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.legend(title='Predicted Species', loc='upper right', bbox_to_anchor=(1.5, 1))
plt.show()
```

```python
def plot_incorrect_predictions(df, x_axis_feature, y_axis_feature):
    fig, ax = plt.subplots(2,2, figsize=(10,10))
    ax = ax.flatten()
    sns.scatterplot(data=df, x=x_axis_feature, y=y_axis_feature, hue="prediction_label", palette='Set1', ax=ax[0])
    sns.scatterplot(data=df, x=x_axis_feature, y=y_axis_feature, hue="target_name", palette='Set1', ax=ax[1])
    sns.scatterplot(data=df, x=x_axis_feature, y=y_axis_feature, hue="correctly_predicted", palette='Set1', ax=ax[2])
    ax[3].set_visible(False)
```

```python
plot_incorrect_predictions(df_predictions, "petal length (cm)", "petal width (cm)")
```
